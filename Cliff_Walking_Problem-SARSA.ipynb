```python
env = gym.make("CliffWalking-v1")
```

```python
print(env.observation_space.n) #Number of state
print(env.action_space.n) # number of action

state,_ = env.reset() #36=>12*3+0
print(state) # starting state
```

### SARSA

```python
gamma    =0.99
alpha    = 0.5
epsilon  = 0.1
episodes = 1000
```

```python
#Q-table => Store Q-Values

Q = np.zeros((48,4))
```

```python
# Policy - epsilon-greedy : state -> action

def epsilon_greedy(state):
    if random.random()<epsilon:
        return env.action_space.sample() #Random Action Explore
    else:
        return np.argmax(Q[state]) #Exploit
```

```python
for episode in range(episodes):
    env = gym.make("CliffWalking-v1")

    done=False #When epsidoe has ended
    state,_ = env.reset()
    action = epsilon_greedy(state)

    total_reward =0
    episode_len=0

    while not done:
        next_state,reward,terminated,truncated,_ = env.step(action)
        done = terminated or truncated
        next_action = epsilon_greedy(next_state)

        #Sarsa update
        Q[state,action] += alpha*(reward + gamma*(Q[next_state,next_action]-Q[state,action]))

        state = next_state
        action = next_action
        total_reward+=reward
        episode_len+=1
    print(f"episode = {episode+1}/500 tot reward = {total_reward} & episode length = {episode_len}")
    env.close()

```
